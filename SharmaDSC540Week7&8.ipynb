##### Split/Apply/Combine

```python
# Finding the oldest and most recent object within each city for each `Medium` using using split-apply-combine
# This function takes in dataframe and a column name, then return the minimum and maximum value as pandas series of that column
def oldest_and_most_recent(df, column_name="AccessionYear"):
    min_value = df[column_name].min()
    max_value = df[column_name].max()
    return pd.Series([max_value,min_value], index=["recent","oldest"])
```

```python
# Create grouped object
grouped_year = merged_df.groupby(["City","Medium"])
```

```python
# Apply the the user defined function to the grouped object
grouped_year.apply(oldest_and_most_recent)
```

### Chapter 11

##### Frequencies and date offsets

```python
# Lets create a subset of dataset with columns for timeseries operations. Using columns `Object Number, AccessionYear, Object Name,Title, Medium`
df_time = df[['Object Number',
              'AccessionYear',
              'Object Name',
              'Title',
              'Medium',
             ]]
```

```python
# view few rows
df_time.head()
```

```python
# Lets convert `AccessionYear` to date timestamp defaulting to `January 1st for date of every year`
accession_year_ts = pd.to_datetime(df["AccessionYear"],format="%Y")
```

```python
# view the timeseries
accession_year_ts.head()
```

```python
# Make the AccessionYear series as the index of the dataframe i.e time series indexed dataframe
df_time.index = accession_year_ts
```

```python
# View few rows
df_time.head()
```

```python
from pandas.tseries import offsets

# Offset the timeseries to 12 noon using Hour offset
accession_year_ts = accession_year_ts + offsets.Hour(12)
```

```python
# Set dataframe index to the 12 noon ofsetted time series
df_time.index = accession_year_ts
```

```python
#view few rows
df_time.head()
```

##### Convert timestamps to periods and back

```python
# convert above dataframe with index as Timestamp to Yearly periods
df_yearly = df_time.copy() #copy dataframe
df_yearly.index = df_time.index.to_period("Y")
```

```python
# view few rows
df_yearly.head()
```

```python
# Converting back to Timestamp
df_timestamp = df_yearly.copy() #copy dataframe
df_timestamp.index = df_yearly.index.to_timestamp(how="start")
```

```python
# view few rows after conversion
df_timestamp.head()
```

##### Period Frequency conversions

```python
# As we have duplicate year in period index, lets drop duplicate index as we will be sampling to change period frequency
df_yearly = df_yearly.groupby(df_yearly.index).first()
```

```python
# Upsampling to `quarter level` and back filling data
df_yearly = df_yearly.resample("Q").ffill()
```
